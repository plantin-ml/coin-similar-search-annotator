```python
import os

import fiftyone as fo
import fiftyone.brain as fob
import fiftyone.zoo as foz
import pandas as pd
from tqdm.auto import tqdm
import numpy as np

ROOT_IMG_DIR = '/home/gpubox2/datasets/coins-data'
```

```python
data_df = pd.read_csv('/home/gpubox2/plantin-projects/coins/similar-search-annotator/api-service/usa-coin-completed.csv')

fo_dataset = fo.Dataset('exp-usa-obverse', persistent=True, overwrite=True)
```

```python
data_df.iloc[0]
```

```python
_data_samples = []
for idx, row in tqdm(data_df.iterrows()):  
    sample = fo.Sample(filepath=os.path.join(ROOT_IMG_DIR, row['file_name']))
    sample['task_alias'] = row['task_alias']
    
    _data_samples.append(sample)
    
fo_dataset.add_samples(_data_samples)
    
```
